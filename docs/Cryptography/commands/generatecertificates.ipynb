# Generate keys in certificate format Commands

````polyglot-notebook
dotnet tool install --global Fhi.Security.Cryptography.CLI --prerelease

$certificateCommonName = "CertName"
$certificatePassword = "CertPw123!"
$certificateDirectory = "C:\Certificates\test1"


fhi-cryptography generatecertificate --CertificateCommonName $certificateCommonName --CertificatePassword $certificatePassword --CertificateDirectory $certificateDirectory
````

## Command
```
 fhi-cryptography generatecertificate --CertificateCommonName <NAME> --CertificatePassword <PASSWORD> --CertificateDirectory <PATH> [--ValidityMonths <MONTHS>]
```

By default, the certificate is valid for **24 months**. Use `ValidityMonths` to customize the validity period.

## Parameters

| Parameter name        | Short name| Description																					| Required | Sample						|
|-----------------------|----------| -----------------------------------------------------------------------------------------------|----------|-------------------------------|
| CertificateCommonName |cn | Common name of the certificate, key file name prefix. <br><br>Files generated: <br>`<CommonName>_private.pfx`<br> `<CommonName>_public.pem`<br>  `<CommonName>_thumbprint.txt`|<b>Yes</b>|`"newCertificate"`|
| CertificatePassword   |pwd |Password for the private certificate.|<b>Yes</b>|`"CertPassword123!"`|
| CertificateDirectory  |dir |Path to where private and public key will be stored.|<b>No</b>|`"C:\\temp"`|
| ValidityMonths        |vm  |Number of months the certificate is valid. Default: 24|<b>No</b>|`18`|
