# Generate new Json Web keys Command

````polyglot-notebook
dotnet tool install --global Fhi.Security.Cryptography.CLI --version 1.0.0-beta1

$keyFileName = ""
$keyDirectory = ""
$keyCustomKid = ""

# Generate keys as JSON files (default)
fhi-cryptography generatejsonwebkey --KeyFileNamePrefix $keyFileName --KeyDirectory $keyDirectory --KeyCustomKid $keyCustomKid

# Generate keys as base64-encoded content
fhi-cryptography generatejsonwebkey --KeyFileNamePrefix $keyFileName --KeyDirectory $keyDirectory --OutputFormat base64
````

## Command
```
 generatejsonwebkey --KeyFileNamePrefix <NAME> --KeyDirectory <PATH> --KeyCustomKid <CUSTOMKID> --OutputFormat <FORMAT>
```

## Parameters

|Parameter name | Short Name| Description																					                                                                                                       | Required | Sample						|
|---------------|------------------------|----------------------------------------------------------------------------------------------------------------------------------------|----------|-------------------------------|
|--KeyFileNamePrefix|-n| Prefix of name of the public and private key file. <br> The keys will be named `<FileName>_private.json` and `<FileName>_public.json`. |<b>Yes</b>|`"newKey"`|
|--KeyDirectory|-d| Path to where private and public key will be stored.                                                                                   |No|`"C:\\temp"`|
|--KeyCustomKid|-k| Optional value to set a custom Kid inside the keys.                                                                                   |No|`"1234567890"`|
|--OutputFormat|-of| Output format: `json` (default) or `base64` for base64-encoded content. When `base64`, files are saved as `.txt` instead of `.json`.  |No|`"base64"`|
